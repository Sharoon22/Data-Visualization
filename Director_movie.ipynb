```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Unnamed: 0: Likely an index column or a redundant ID column.
# id: The unique identifier for each movie or data entry.
# budget: The allocated budget for the movie, likely in millions.
# popularity: A measure of how popular the movie is, based on factors like views, mentions, etc.
# revenue: The total earnings or revenue the movie generated, possibly in millions.
# title: The title or name of the movie.
# vote_average: The average rating the movie received, usually out of 10 (scaled to 100 here).
# vote_count: The number of people who voted or rated the movie.
# director_id: The unique identifier for the movie's director.
# year: The year the movie was released or produced.
# month: The month the movie was released or produced.
# day: The day of the month the movie was released or produced.
```

```python
movies=pd.read_csv('/Users/sharoonkk/Desktop/DA/Matplotib/movies.csv')
```

```python
movdire=movies.merge(directors,how='left',left_on='director_id', right_on='id',indicator=True)
```

```python
movdire
```

```python
movdire.isnull().sum()
```

```python
movd=movdire.drop(columns=['Unnamed: 0_y','id_y'],axis=1)
```

```python
movd.dropna(inplace=True)
```

```python
movd.isnull().sum()
```

```python
movd.rename(columns={'Unnamed: 0_x': 'Unnamed: 0', 'id_x': 'id'}, inplace=True)
```

```python
movd.duplicated().sum()
```

```python
movd.describe()
```

```python
movd
```

## 1 The movies that was directed by Robert Rodriguez and how many revenues generated

```python
movdir=movdire[movdire['director_name']=='Robert Rodriguez'][['title','director_name','revenue']]
```

```python
movdir
```

## 2 The movie that released on 2009 and who was the director and how much is the budget of that movie

```python
movie=movd[movd['year']==2009][['title','director_name','budget','year']]
```

```python
movie
```

```python
plt.figure(figsize=(10,20))
plt.plot(movie['budget'],movie['director_name'],marker='d')
plt.ylabel('year')
plt.xlabel('budget')
plt.title('movies')
plt.grid(True)
plt.show()
```

## 3 What was the average revenue generated by Robert Rodriguez

## 4 Which movie generated highest revenue how much was it and in which year the movie was released and who was the director

```python
maxrevenue = movd.loc[[movd['revenue'].idxmax()], ['title','revenue','year','director_name']]
```

```python
maxrevenue
```

## 13 List the top 10 directors who earn more revenues from all their movies
